```
from keras import models
from keras import layers
```

```
network = models.Sequential()
```

1. Transforma a matrix 28 * 28 em uma lista com 784 linhas que será e entrada da rede neural
2. O segundo layer é composto por 512 neuronios
3. A saída é enviada para uma camada com 10 neurônios (softmax saída 0 ou 1)

```
network.add(layers.Dense(512, activation='relu', input_shape=(28*28,)))
network.add(layers.Dense(10,activation='softmax'))
```

```
network.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```
from keras.utils import to_categorical
```

```
train_labels.shape
```

```
train_labels2 = to_categorical(train_labels)
test_labels2  = to_categorical(test_labels)
```

```
test_labels2.shape
```

```
train_labels[0]
```

```
train_labels2[0]
```

```
train_images2 = train_images.reshape(60000,28*28)
train_images2.shape
```

```
test_images2 = test_images.reshape(10000, 28*28)
test_images2.shape
```

Vamos normalizar os valores de entrada entre 0 e 1

Valor normalizado = (Valor Orginal - Menor Valor) / (Maior Valor)

Neste caso, como  Menor Valor = 0 e Maior Valor = 255, normalizar significa dividir por 25.



```
train_labels2.shape
```

```
train_images3 = train_images2 / 255
test_images3 = test_images2 / 255

train_labels3 = train_labels2
test_labels3 = test_labels2
```

No exemplo 60.000 / 128 = 468  --> 10 epocas significa 10 8 468

```
network.fit(train_images3, train_labels3, epochs=50, batch_size=128)
```

```
test_images3.shape
```

```
test_labels3.shape
```

```
test_loss, test_acc = network.evaluate(test_images3, test_labels3)
```
